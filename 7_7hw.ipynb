##Gitの基本的な使い方
#インストール
まずはGitをインストールします。公式サイトからダウンロードできます。

#初期設定
Gitを初めて使う際には、ユーザー名とメールアドレスを設定する必要があります。


```
git config --global user.name "Your Name"
git config --global user.email "your.email@example.com"
```


#リポジトリの作成
新しいプロジェクトを開始する際には、新しいリポジトリを作成します。


```
mkdir myproject
cd myproject
git init
```


#既存のリポジトリをクローン
既存のリポジトリをクローン（複製）することで、そのプロジェクトの全ファイルを取得できます。


```
git clone https://github.com/username/repository.git
```


#変更の記録
変更をステージングエリアに追加し、コミットしてリポジトリに記録します。


```
git add filename
git commit -m "コミットメッセージ"
```
すべての変更を追加する場合は、git add .を使用します。

#履歴の確認
リポジトリの変更履歴を確認します。


```
git log
```


#ブランチの操作
新しいブランチを作成し、移動します。


```
git branch new-branch
git checkout new-branch
```
または、git checkout -b new-branchで同時に作成と移動を行えます。

#マージ
ブランチをマスター（main）ブランチにマージします。


```
git checkout main
git merge new-branch
```


#リモートリポジトリの操作
リモートリポジトリに変更をプッシュします。


```
git push origin main
```


リモートリポジトリの変更をフェッチし、ローカルリポジトリを更新します。


```
git fetch
git pull
```

##GitHubの基本的な使い方
#アカウントの作成
GitHubの公式サイトでアカウントを作成します。

#リポジトリの作成
新しいリポジトリを作成するには、GitHubのダッシュボードから「New Repository」をクリックし、必要な情報を入力します。

#リモートリポジトリの設定
ローカルリポジトリをGitHubのリモートリポジトリに関連付けます。


```
git remote add origin https://github.com/username/repository.git
```


#リモートリポジトリへのプッシュ
ローカルの変更をリモートリポジトリにプッシュします。


```
git push -u origin main
```


#プルリクエストの作成
他の人のリポジトリに変更を提案するには、フォークして変更を加えた後、プルリクエストを作成します。

#プッシュする際の認証について
GitHubでは2021年8月13日にパスワード認証が廃止されました。<br>
パスワードではなくアクセストークン認証によってGitHubへプッシュする必要があります。
GitHubにログインした状態で、以下を実施します。

 - 設定画面( https://github.com/settings/tokens )を開く
 - [Generate new token]をクリック
 - noteに任意の名前を記入
 - 権限を設定
 - [Generate token] をクリック
 - トークンをコピー　(ページを離れると参照不可になる)
 - パスワードの代わりに使用する

注意：権限設定で[repo] を選択します。<br>
